```python
import os
from PIL import Image, ImageOps
from fpdf import FPDF

# Paths
output_folder = r"C:/Users/mered.mulugeta.ETHIO.000/Desktop/Barcode/Test5"

# PDF creation function
def create_pdf_from_images(image_paths, pdf_path):
    pdf = FPDF()
    pdf.set_auto_page_break(0)
    
    # Dimensions for each ticket and space
    ticket_image_width = 100  # Image width in mm
    ticket_image_height = 50  # Image height in mm
    spacing_x = 5  # Horizontal space between tickets in mm
    spacing_y = 5   # Vertical space between tickets in mm
    
    tickets_per_row = 2  # 2 tickets per row
    tickets_per_column = 5  # 5 tickets per column (10 tickets per page)

    pdf.add_page()  # Start with the first page

    for index, image_path in enumerate(image_paths):
        # If we've placed 10 tickets, start a new page
        if index > 0 and index % (tickets_per_row * tickets_per_column) == 0:
            pdf.output(pdf_path)  # Save the current PDF
            pdf = FPDF()
            pdf.set_auto_page_break(0)
            pdf.add_page()  # Add a new page for the next set of 10 tickets
            pdf_path = pdf_path.replace(".pdf", f"_{index // (tickets_per_row * tickets_per_column) + 1}.pdf")
        
        img = Image.open(image_path)
        img = ImageOps.fit(img, (int(ticket_image_width * 3.78), int(ticket_image_height * 3.78)), method=Image.Resampling.LANCZOS)
        
        x = (index % tickets_per_row) * (ticket_image_width + spacing_x) + 10  # Adjust the horizontal positioning
        y = (index // tickets_per_row % tickets_per_column) * (ticket_image_height + spacing_y) + 10  # Adjust the vertical positioning
        
        pdf.image(image_path, x=x, y=y, w=ticket_image_width, h=ticket_image_height)

    pdf.output(pdf_path)  # Output the last PDF

# Generate 100 PDFs with 10 tickets each
for batch in range(1, 101):
    start_index = (batch - 1) * 10
    end_index = start_index + 10
    ticket_images = [os.path.join(output_folder, f"ticket+1614{str(i).zfill(4)}.png") for i in range(start_index + 1, end_index + 1)]
    
    pdf_output_path = os.path.join(output_folder, f"tickets_batch_{batch}.pdf")
    
    create_pdf_from_images(ticket_images, pdf_output_path)

print("100 PDF files with 10 tickets each generated successfully.")
```
